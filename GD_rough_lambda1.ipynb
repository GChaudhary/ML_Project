```python
fileName = 'F:/Machine_Learning/MATLAB/impData/First_Observation_fix2.npz'
```

```python
print ('Loading observations...\n')

import numpy as np
observations = np.load(fileName)

Jhistory = observations['Jhistory']
Accuracy = observations['Accuracy']
Alpha = observations['Alpha']
Iters = observations['Iters']


print 'Jhistory -->', Jhistory.shape
print 'Accuracy -->', Accuracy.shape
print 'Alpha    -->', Alpha.shape
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(Alpha, 'ro')
plt.xlabel('Array Index')
plt.ylabel('Alpha')
plt.title("Different values of 'alpha'")
```

```python
plt.figure(figsize=(5, 5))
plt.subplots_adjust(hspace=0.7)

plt.subplot(2, 1, 1)
plt.plot(Alpha, Accuracy[:, 0], 'ko-')
plt.title('TRAINING Set Accuracy')
plt.xlabel('Alpha')
plt.ylabel('Accuracy (%)')

plt.subplot(2, 1, 2)
plt.plot(Alpha, Accuracy[:, 1], 'ko-')
plt.title('TEST Set Accuracy')
plt.xlabel('Alpha')
plt.ylabel('Accuracy (%)')

plt.show()
```

```python
plt.figure(figsize=(7, 21))
for i in range(14):    
    plt.subplot(7, 2, i+1)
    plt.plot(Jhistory[i, :], 'r')
    str_title = "Alpha = " + str(Alpha[i, 0]) + "\nMin Cost = " + str(Jhistory[i, -1])
    plt.title(str_title)
    plt.ylabel('Cost')
    plt.xlabel('# of iterations')
    plt.subplots_adjust(left=None, bottom=None, wspace=0.6, hspace=0.7)

plt.show()
```
